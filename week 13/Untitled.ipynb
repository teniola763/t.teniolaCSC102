```python
import pandas as pd 
print("Welcome to pau-sims")
class sims():
     
    
    
    #attributes
    def __init__(self):
        
        self.name = []
        self.matric = []
        self.dept = []
        self.level = []
                   


    def folder(self):
                    
        folder = {}
        folder["student Name"] = self.name
        folder["level"] = self.level
        folder["matric"] = self.matric
        folder["department"] = self.dept

        df = pd.DataFrame(folder)
        df.to_csv("Sims.csv", index_label= "No", na_rep = "M/A")
        
    def success(self):
        print('file has been created successfully')


    def inps(self):
        
            names = input("do please enter your name: ")
            self.name.append(names)
            matric = input("do please  enter your matric number: ")
            self.matric.append(matric)
            department = input("do please enter your department: ")
            self.dept.append(department)
            level = input("do please enter your level: ")
            self.level.append(level)

ask = int(input('Input number of students: '))                            
obj = sims()

for x in range(ask):
    obj.inps()
    obj.folder()

obj.success()
    





```
